```python
reviews.iloc[115]['translation'] = ""
```

```python
for index, review in reviews.iterrows():
    body = review['body']
    trans = review['translation']
    
    #print(trans)
    #print(type(trans))

    lang = detect(body)
    #print(lang)
    reviews.iloc[index]['lang'] = lang
    #print(body, '\n')

    text_area = driver.find_element_by_class_name('form-control')

    if lang != 'en' and (type(trans) == np.float or trans == ""):
        try:
            driver.find_element_by_class_name("form-control").send_keys( body )
            time.sleep(14)

            content = driver.page_source
            soup = BeautifulSoup(content)    

            translation_ = soup.find(id="outputEditor")
            translation = translation_.text
            reviews.iloc[index]['translation'] = translation

            #print(body,'\n')
            #print(translation,'\n')
            for i in range(len(body)):
                text_area.send_keys(Keys.BACKSPACE)
            time.sleep(8)
        except: 
            for i in range(len(body)):
                text_area.send_keys(Keys.BACKSPACE)
            print('skip',index)
            time.sleep(8)
    else:
        print('not searched',index)
        if (type(trans) != np.float or trans == ""):
            reviews.iloc[index]['translation'] = trans
            pass
        else:    
            translation = body
            reviews.iloc[index]['translation'] = translation
    
    #print(translation, '\n')
    print(index)
```
